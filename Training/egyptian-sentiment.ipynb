# Preprocessing


## 1-Text Cleaning

```python
def clean_text(text):
    text = re.sub(r'[\u064B-\u0652]', '', text)
    text = re.sub(r'[0-9Ù -Ù©]', '', text)
    text = re.sub(r'[^\w\s]', ' ', text)
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'(.)\1+', r'\1', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

## 2- Normalization

## In Arbic Some Letters Can Be in Diffrent Form I will Handle it

```python
def normalize_arabic(text):
    text = re.sub("[Ø¥Ø£Ø¢Ø§]", "Ø§", text)
    text = re.sub("Ù‰", "ÙŠ", text)
    text = re.sub("Ø¤", "Ùˆ", text)
    text = re.sub("Ø¦", "ÙŠ", text)
    text = re.sub("Ø©", "Ù‡", text)
    return text
```

## 3-Remove Stop Words

```python
nltk.download('stopwords')
nltk.download('punkt')
```

```python
arabic_stopwords = set(stopwords.words('arabic'))

def remove_stopwords(text):
    words = word_tokenize(text)
    words = [w for w in words if w not in arabic_stopwords]
    return " ".join(words)
```

## 5-Lemmatization

```python
lemmatizer = qalsadi.lemmatizer.Lemmatizer()

def lemmatize_text(text):
    words = word_tokenize(text)
    lemmas = [lemmatizer.lemmatize(word) for word in words]
    return " ".join(lemmas)
```

```python
def preprocess_arabic_text(text):
    text = clean_text(text)
    text = normalize_arabic(text)
    text = remove_stopwords(text)
    text = lemmatize_text(text)
    return text
```

```python
df["clean_tweet"] = df["tweet"].apply(preprocess_arabic_text)
```

```python
class ArabicTweetDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_length=128):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_length = max_length
    
    def __len__(self):
        return len(self.texts)
    
    def __getitem__(self, idx):
        text = str(self.texts.iloc[idx]) if hasattr(self.texts, 'iloc') else str(self.texts[idx])
        label = self.labels.iloc[idx] if hasattr(self.labels, 'iloc') else self.labels[idx]
        
        encoding = self.tokenizer(
            text,
            truncation=True,
            padding='max_length',
            max_length=self.max_length,
            return_tensors='pt'
        )
        
        return {
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(label, dtype=torch.long)
        }

```
